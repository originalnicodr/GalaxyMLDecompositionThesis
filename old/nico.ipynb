```python
import galaxychop as gchop



gal = gchop.preproc.center_and_align(gchop.io.read_hdf5("tests/datasets/galaxy_TNG_611399.h5"))

comp = gchop.models.JHistogram().decompose(gal)

```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df, hue = gal.plot.get_df_and_hue(None, ["x", "y", "z"], comp, lmap=None)

fig, axs = plt.subplots(5, 3, figsize=(6, 2*5), sharex=True, sharey=True)

axs[0,0].set_ylabel("Abadi")
sns.histplot(x="x", hue=hue, data=df, ax=axs[0,0], legend=False)
sns.histplot(x="y", hue=hue, data=df, ax=axs[0,1], legend=False)
sns.histplot(x="z", hue=hue, data=df, ax=axs[0,2], legend=False)


axs[1,0].set_ylabel("Single")
...

axs[2,0].set_ylabel("AVG")
...

axs[3,0].set_ylabel("Complete")
...

axs[4,0].set_ylabel("Ward")
...


axs[-1,0].set_xlabel("X")
axs[-1,1].set_xlabel("Y")
axs[-1,2].set_xlabel("Z")

fig.tight_layout();
```

```python
df, hue = gal.plot.get_df_and_hue(None, ["x", "y", "z"], comp, lmap=None)

fig, axs = plt.subplots(5, 3, figsize=(6, 2*5), sharex=True, sharey=True)

axs[0,0].set_ylabel("Abadi")
sns.histplot(x="x", y="y", hue=hue, data=df, ax=axs[0,0], legend=False)
sns.histplot(x="y", y="z", hue=hue, data=df, ax=axs[0,1], legend=False)
sns.histplot(x="x", y="z", hue=hue, data=df, ax=axs[0,2], legend=False)

axs[-1,0].set_xlabel("XY")
axs[-1,1].set_xlabel("YZ")
axs[-1,2].set_xlabel("XZ")

fig.tight_layout();
```
